```python
Write a function which will accept 2 numbers from used and return their division, floor division, modulus 
```

```python
def division():
    a=int(input("enter a number"))
    b=int(input("enter a number"))
    d = a / b
    f = a // b
    m = a % b   
    print("division",d)
    print("floor",f)
    print("modulus",m)
division()   
```
